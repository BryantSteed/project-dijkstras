```python
from graph_generation_stretch2 import get_plane_based_graph
from network_routing import find_shortest_path_with_heap
import matplotlib.pyplot as plt
import random
from plotting import plot_points, plot_weights, show_plot, draw_path, circle_point
```

```python
seed = random.randint(0, 10000)
small_graph, small_plane = get_plane_based_graph(seed, 100, 0.1, 10)
medium_graph, medium_plane = get_plane_based_graph(seed, 100, 0.2, 10)
large_graph, large_plane = get_plane_based_graph(seed, 100, 0.3, 10)
```

```python
def display_graph(graph: dict[int, dict[int, float]], point_plane: list[tuple[float, float]],) -> None:
    path, cost = find_shortest_path_with_heap(graph, 0, 99)
    plot_points(point_plane)
    plot_weights(point_plane, graph)
    draw_path(point_plane, path)
    circle_point(point_plane[0])  # start point
    circle_point(point_plane[99])  # end point
    show_plot()
```

```python
display_graph(small_graph, small_plane)
display_graph(medium_graph, medium_plane)
display_graph(large_graph, large_plane)
```
