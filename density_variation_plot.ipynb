```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([500, 1000, 1500, 2000, 2500, 3000, 3500])
linear_low_density = np.array([18, 80, 180, 331, 573, 1240, 1207])
heap_low_density = np.array([11, 47, 87, 153, 255, 399, 534])

linear_medium_density = np.array([25, 103, 243, 452, 764, 1170, 1789])
heap_medium_density = np.array([16, 63, 138, 259, 434, 668, 931])

linear_high_density = np.array([31, 130, 315, 593, 1039, 1703, 2494])
heap_high_density = np.array([25, 95, 225, 394, 659, 1031, 1672])
```

```python
fig, axes = plt.subplots(3, 1, figsize=(7, 9), sharex=True, layout='constrained')
fig.suptitle("Dijkstra's Algorithm: Linear vs Heap Priority Queue\nVarying Graph Density", fontsize=16)

axes[0].plot(x, linear_low_density, marker='o', label='Linear')
axes[0].plot(x, heap_low_density, marker='o', label='Heap')
axes[0].set_title("Low Density 0.3")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(x, linear_medium_density, marker='o', label='Linear')
axes[1].plot(x, heap_medium_density, marker='o', label='Heap')
axes[1].set_title("Medium Density 0.5")
axes[1].legend()
axes[1].grid(True)

axes[2].plot(x, linear_high_density, marker='o', label='Linear')
axes[2].plot(x, heap_high_density, marker='o', label='Heap')
axes[2].set_title("High Density 1.0")
axes[2].legend()
axes[2].grid(True)
axes[2].set_xlabel("Number of Nodes (V)")
for a in axes: a.set_ylabel("Time (ms)")

plt.show()
```
